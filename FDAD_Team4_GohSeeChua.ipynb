## Data Cleaning


We will replace all values in Fuel Type with the actual name of each Fuel Type for more clear visualisation.

```python
cd_data["Fuel Type"].replace('X', 'Regular Gasoline', inplace=True)
cd_data["Fuel Type"].replace('Z', 'Premium Gasoline', inplace=True)
cd_data["Fuel Type"].replace('D', 'Diesel', inplace=True)
cd_data["Fuel Type"].replace('E', 'Ethanol', inplace=True)
cd_data["Fuel Type"].replace('N', 'Natural Gas', inplace=True)
```

We are interested in finding out the relationship between Transmission Type and C02 Emissions. However, the Transmission Type categorical variable from our dataset has 2 aspects- firstly, the letters (e.g. A, AM etc) which stand for the Transmission Type. Secondly, some of the data have a number next to the letters, which stands for the number of gears each Transmission Type uses. However, same Transmission Types can have different number of gears, e.g. AS5 and AS8. We will create a new column "Gears" to do an analysis to determine if the number of gears is a significant variable in determining the CO2 Emissions(g/km). From our data description file, number of gears ranges from 3 - 10.

```python
cd_data["Gears"] = None

for row in range(len(cd_data)):
    for character in cd_data.iloc[row, 5]:
        if character.isdigit() and character == '1':
            cd_data.at[row, "Gears"] = 10
            break
        elif character.isdigit():
            cd_data.at[row, "Gears"] = int(character)
cd_data
```

```python
print("Data type: ", type(cd_data))
print("Data dims: ", cd_data.shape)
```

```python
print(cd_data.dtypes)
```

Checking for NULL values in the new "Gears" column

```python
cd_data.isnull().sum()
```

Based on our research, the rows with NULL value in "Gears" column are the vehicles with an AV transmission (i.e. continuously variable transmission). These cars do not have a set number of gears as compared to other normal automatic transmission cars, but instead has 1 special gear that is variable for all driving conditions. So, to clean up the dataset and to ensure each row has a corresponding Gear value, we will replace all NULL values with the value 1. 

```python
cd_data["Gears"].fillna(value=1, inplace=True)
```

Now that the "Gears" column has been cleaned up, we will do an analysis on the "Gears" variable with respect to the "CO2 Emissions (g/km)" variable, to determine if the number of gears is a significant variable in predicting the amount of CO2 emissions of a car. Also note that Gears is a categorical variable- each number of gears has to be a discrete integer.

```python
sb.catplot(data=cd_data, x="Gears", y="CO2 Emissions(g/km)", kind='box')
```

From the boxplot, we can conclude that generally speaking (except for 4 and 9 Gears), judging by the median of each plot, as number of Gears increases, the CO2 Emissions increase as well. There is a relationship between Gears and CO2 Emissions, and we will therefore include it in our data analysis further below.

We will also create a new Transmission Type column (A, AM, AS, AV, M) to do a separate analysis using the Transmission Type of the car. 

```python
cd_data["Transmission Type"] = None
for row in range(len(cd_data)):
    for characters in cd_data.iloc[row, 5]:
        if characters.isalpha():
            if cd_data.at[row, "Transmission Type"] == None:
                cd_data.at[row, "Transmission Type"] = characters
            else:
                cd_data.at[row, "Transmission Type"] += characters
        elif characters.isdigit():
            break
cd_data
```

# Investigating the multicollinearity of the 5 categorical variables

```python
from scipy.stats import chi2_contingency
import itertools

def generate_pairs(arr):
    return list(itertools.combinations(arr, 2))

def apply_function_to_pairs(arr):
    pairs = generate_pairs(arr)

    for x, y in pairs:
        chi2, p_value, dof, ex = chi2_contingency(pd.crosstab(cd_data[x], cd_data[y]))
        print(f"p value for chi square test for {x} and {y} is: {p_value}")
    return

apply_function_to_pairs(["Cylinders", "Fuel Type", "Transmission Type", "Vehicle Class", "Gears"])
```

Our null hypothesis is that each categorical variable is independent from every other categorical variable. For all (5C2 == 10) cases, our chi squared test returned a p-value that is very close to 0, or 0. Using a 95% confidence test, we found that p-value for all cases was much smaller than the alpha level of 0.05, so we do not reject the null hypothesis. We can say that all the categorical variables are independent from each other- there is no multicollinearity.

### One-hot encoding for the categorical variables:

We will convert the categorical variables of the dataset into numerical variables to be used in our Regression Models. 

```python
onehotencodingdata = cd_data.copy()
encoded_data = ['Fuel Type', 'Vehicle Class', 'Transmission Type']
encoded_data_prefix = ['FT', 'VClass', 'TType']
for col, per in zip(encoded_data, encoded_data_prefix):
  dummies = pd.get_dummies(onehotencodingdata[col], prefix=per, drop_first=False)
  onehotencodingdata = onehotencodingdata.join(dummies)

removable_data = encoded_data + ['Make', 'Model','Fuel Consumption City (L/100 km)','Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (mpg)', 'CO2 Emissions(g/km)', 'Transmission']

removable_data
```

```python
final_data = onehotencodingdata.drop(removable_data, axis=1)
final_data.head(20)
```

# Creating a regression model to predict emissions of a vehicle 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = final_data
```

```python
Y = pd.DataFrame(cd_data["CO2 Emissions(g/km)"])
```

```python
linreg = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.20, random_state = 69)
print("Train Set :", y_train.shape, X_train.shape)
print("Test Set  :", y_test.shape, X_test.shape)
```

```python
X_train.describe()
```

```python
y_train.describe()
```

### Model 1: Multilinear Regression Model

```python
#we fit our train data onto our multilinear regression model
linreg.fit(X_train, y_train)
```

```python
print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_)
```

```python
y_train_pred = linreg.predict(X_train)
y_test_pred = linreg.predict(X_test)

# Plot the Predictions vs the True values
f, axes = plt.subplots(1, 2, figsize=(24, 12))
axes[0].scatter(y_train, y_train_pred, color = "blue")
axes[0].plot(y_train, y_train, 'w-', linewidth = 1)
axes[0].set_xlabel("True values of the Response Variable (Train)")
axes[0].set_ylabel("Predicted values of the Response Variable (Train)")
axes[1].scatter(y_test, y_test_pred, color = "green")
axes[1].plot(y_test, y_test, 'w-', linewidth = 1)
axes[1].set_xlabel("True values of the Response Variable (Test)")
axes[1].set_ylabel("Predicted values of the Response Variable (Test)")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

y_train_pred = linreg.predict(X_train)

print("Metrics of the Multilinear Regression Model on Train Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_train, y_train))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_train, y_train_pred))
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mean_squared_error(y_train, y_train_pred)))

y_test_pred = linreg.predict(X_test)
```

```python
print("Metrics of the Multilinear Regression Model on Test Dataset")
print("Explained Variance (R^2) \t:", linreg.score(X_test, y_test))
print("Mean Squared Error (MSE) \t:", mean_squared_error(y_test, y_test_pred))
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
StdE_pred = np.sqrt(len(y_train) * mean_squared_error(y_train, y_train_pred) /(len(y_train) - 2))
print("Standard Error of Prediction \t:",StdE_pred)
print()
```

```python
# to check the original data's CO2 Emissions mean to serve as a rough sense of scale when deciding if RMSE values are significant.
print("The mean of CO2 Emisisons: ", cd_data["CO2 Emissions(g/km)"].mean())
```

Comparing R^2 values for both Train and Test datasets, we see that the Explained Variance for both are very close to the maximum of 1, which suggests that the model has an excellent goodness of fit. Comparing the RMSE values for both datasets, again they are very similar, and approximately 50 times less than the mean of the CO2 Emissions predictor variable. This suggests that RMSE for both datasets is not significant compared to CO2 Emissions. So, the prediction accuracy of the model is also very good.

# Evaluating the best regression model


Firstly, we compare the statistics of our 3 models. The R^2 values for all 3 models are all very close to the maximum of 1, which suggests excellent goodness of fit for all 3 models. The Root Mean Square Error values are again very similar, and relatively much smaller compared to the scale of our data. Again, this suggests that RMSE is not significant, which suggests that prediction accuracy of all 3 models is excellent.

However, to evaluate the best performing model, we have to consider certain other metrics. We note that as compared to the other 2 models, the Multilinear Regression Model generates the coefficients for each variable rather than for each level of the variable. This makes it difficult for us to analyse the impact of the different specifications of a car on CO2 Emissions. 


Comparing between Lasso and Ridge, we think that Lasso Regression is a better model because it is able to produce a simpler model since it performs automatic feature selection by shrinking non-important variables to 0. Lasso Regression also helps us to better identify the most significant variables in impacting CO2 Emissions, having a similar performance to Ridge Regression in terms of R^2 and MSE, therefore we chose Lasso Regression to help us in our recommendations.
